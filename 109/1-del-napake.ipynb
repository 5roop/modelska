```python
from __future__ import print_function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os


np.set_printoptions(precision=4)
plt.style.use("ggplot")
plt.rcParams['font.family'] = "serif"
plt.rcParams['font.size'] = 13
```

```python
masa = pd.read_csv("podatki/masa1.csv")
masa["absraz"] = np.abs(masa.m_hat - masa.m_hat[masa.N.idxmax()])
plt.scatter(masa.N, np.log(masa.absraz))
plt.semilogx()
plt.ylabel(r"$\log_{10} |m - \hat{m}|$")
plt.xlabel("N")
plt.savefig("1-3-napaka_mase.pdf")
plt.show()
```

```python
moment = pd.read_csv("podatki/momentn.csv")
moment["N"] = moment["Unnamed: 0"].apply(lambda row: int(row))

moment["absraz"] = np.abs(moment.m_hat - moment.m_hat[moment.N.idxmax()])
plt.scatter(moment.N, np.log(moment.absraz))
plt.semilogx()
plt.ylabel(r"$\log_{10} |p - \hat{p}|$")
plt.xlabel("N")
plt.savefig("1-3-napaka_vstrajnostnega_momenta.pdf")
plt.show()
```
