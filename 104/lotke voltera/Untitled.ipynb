```python
from __future__ import print_function
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import odeint


plt.style.use("ggplot")
plt.rcParams['font.family'] = "serif"
plt.rcParams['font.size'] = 13
```

```python
def fiddle(p, time, Z0 = 1, L0 = 1):
    def diff_eqs(INP, t):
        Y = np.zeros(2)
        V = INP
        Y[0] = p * V[0] * (1 - V[1])
        Y[1] = V[1] / p *(V[0]-1)    #    Fpika = F/p (A-1)
        return Y
    INPUT = np.array([Z0, L0])
    t_range = time
    RES = odeint(diff_eqs,INPUT, t_range)
    return RES
```

```python
razpon = np.linspace(0,50, 10000)
ZL = fiddle(1, razpon, Z0=20
           )

fig, [ax1, ax2] = plt.subplots(1,2,sharey = True, )
ax1.plot(razpon, ZL[:,0], label="Zajci")
ax1.plot(razpon, ZL[:,1], label="Lisice")
ax1.set_ylabel("Populacija lisic/zajcev")
ax1.set_xlabel("Brezdimenzijski čas")
ax1.legend()


ax2.scatter(ZL[:,0], ZL[:,1])
ax2.set_ylabel("Populacija lisic")
ax2.set_xlabel("Populacija zajcev")
#plt.savefig("1-1-p120Z01,2.pdf")
fig.show()
```
