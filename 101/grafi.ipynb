```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 1, 1000)
plt.style.use("ggplot")
plt.rcParams['font.family'] = "serif"
plt.rcParams['font.size'] = 13
#plt.rcParams.update({'figure.autolayout': True})
plt.rc('text', usetex=False)
```

# Prvi del: imamo obe hitrosti

```python
def obe(t, y0, yk):
    return (3*yk+3*y0-6)*(t**2-2/3 *t)+2*(1-y0)*t+y0
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(121)
for y0 in [0, 0.5, 1, 1.5,2]:
    plt.plot(t, obe(t, y0, 1), label=r"$y_k = 1, y_0 = {}$".format(y0))
    plt.xlabel(r"Brezdimenzijski čas $x$")
    plt.ylabel(r"Brezdimenzijska hitrost $y$")
    plt.legend()
plt.subplot(122)
for y0 in [0, 0.5, 1, 1.5,2]:
    plt.legend()
    plt.plot(t, obe(t, y0, 1.5), label=r"$y_k = 1.5, y_0 = {}$".format(y0))
    plt.xlabel(r"Brezdimenzijski čas $x$")
    plt.ylabel(r"Brezdimenzijska hitrost $y$")

plt.savefig("prvidel.pdf")
plt.savefig("prvidel.png")
plt.show()

```

# Drugi del: končna hitrost poljubna

```python
def nikoncne(t, y0):
    return 3/2*(1-y0)*(2-t)*t+y0
```

```python
plt.figure(figsize=(8,8))
for y0 in [0, 0.5, 1, 1.5,2]:
    plt.legend()
    plt.plot(t, nikoncne(t, y0), label=r"$y_0 = {}$".format(y0))
    plt.xlabel(r"Brezdimenzijski čas $x$")
    plt.ylabel(r"Brezdimenzijska hitrost $y$")


plt.savefig("drugidel.png")
plt.savefig("drugidel.pdf")
plt.show()
```

# Tretji del: poljubne potence pospeška

```python
def tretji(t, y0,n):
    a = 1/(2*n-1)
    odlambde = (  (1-y0)*(a+1))/(- (1-t)**(a+2)/(a+2)-1 )
    return y0 + odlambde/(a+1)*((1-t)**(a+1)-1)
```

```python
plt.figure(figsize=(16,16))
ypsiloni = [0,0.5,1, 1.5]
subploti = [221, 222, 223, 224]
slovar = dict(zip(ypsiloni, subploti))
for y0 in ypsiloni:
    for n in [ 1, 2, 10, 1000]:
        plt.subplot(slovar[y0])
        plt.plot(t, tretji(t, y0, n), label=r"$n = {}, y_0={}$".format(n,y0))
        plt.xlabel(r"Brezdimenzijski čas $x$")
        plt.ylabel(r"Brezdimenzijska hitrost $y$")
        plt.legend()

plt.savefig("tretjidel.png")
plt.savefig("tretjidel.pdf")
plt.show()
```

# Četrti del -- ne vem zakaj mi tukaj ni hotelo zvektorizirat kode -- preveri nekoč

```python
def cetrti(x, y0, b = 0.1):
    try:
        c = cosh(b)
        s = sinh(b)
        return (1-y0) / (s/b-c) * (cosh(b-b*x)-c) + y0
    except ZeroDivisionError:
        pass
ypsiloni = [0,0.5,1, 1.5]

plt.figure(figsize=(12,6))
plt.subplot(121)
for y0 in ypsiloni:
    ylist = [cetrti(i, y0, b = 0.01) for i in t]
    plt.plot(t, ylist, label=r"$y_0 = {}$".format(y0))
    plt.xlabel(r"Brezdimenzijski čas $x$")
    plt.title("b = 0.01")
    plt.ylabel(r"Brezdimenzijska hitrost $y$")
    plt.legend()
plt.subplot(122)
for y0 in ypsiloni:
    ylist = [cetrti(i, y0, b=5) for i in t]
    plt.plot(t, ylist, label=r"$y_0 = {}$".format(y0))
    plt.title("b = 5")
    plt.xlabel(r"Brezdimenzijski čas $x$")
    plt.ylabel(r"Brezdimenzijska hitrost $y$")
    plt.legend()
plt.savefig("cetrtidel.pdf")
plt.savefig("cetrtidel.png")    
plt.show()
```

```python
plt.plot(t, cetrti(t, 0.5), label=r"$y_0 = {}$".format(0.5))
plt.xlabel(r"Brezdimenzijski čas $x$")
plt.ylabel(r"Brezdimenzijska hitrost $y$")
plt.legend()
plt.show()
```

# 5. del

```python
def peti(t,y0):
    A = 6*y0-6
    return A*t**2 - A*t + y0
```

```python
plt.figure(figsize=(8,8))
for y0 in [0,0.5,1,1.5,2]:
    plt.plot(t, peti(t, y0), label=r"$y_0 = {}$".format(y0))
    plt.xlabel(r"Brezdimenzijski čas $x$")
    plt.ylabel(r"Brezdimenzijska hitrost $y$")
    plt.legend(loc="lower center")
plt.savefig("peti.pdf")
plt.savefig("peti.png")
plt.show()
```
